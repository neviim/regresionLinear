```
import numpy as np
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import load_boston
```

```
# compor a biblioteca
boston = load_boston()

# print(boston.DESCR )
```

Formula minimizar o erro quadratico medio (MCO): $\beta =  $\(X^(T)X)^(-1)X^(T)Y$

```
# pega o numero de habitantes em boston, passa a x,y convertido em array
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

# Plota a matriz
plt.scatter(X, Y, alpha=0.4)

#print(X)  // numero de habitação
#print(Y)  // valor medio de venda

# Adicionar uma coluna de 1 para termos independentes
X = np.array([np.ones(506), X]).T # Transposta

#print(X)

# multiplica pela matrix transposta
B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4,9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c="red")
plt.show()
```
